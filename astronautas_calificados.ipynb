
# Crea la columna "promedio_evaluaciones" con el promedio de las 4 columnas
 de evaluaciones

```python
#Creala columna "IMC" usando la fórmula entregada#

df1=pd.DataFrame(df)
df1["altura"]= df1["altura"]/100
df1["IMC"]= df1['peso'] / (df1['altura'] ** 2)
df1["promedio_evaluaciones"]= df[["evaluacion_fisica","evaluacion_psicologica", "evaluacion_adaptativa", "evaluacion_medica"]].mean(axis=1)
df1
```
