```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_cleaned= pd.read_csv("Cleaned_DS_Jobs.csv")
df_cleaned
```

```python
df_cleaned1 = df_cleaned.copy()
print(df_cleaned1)
```

```python
df_cleaned1=pd.DataFrame(df_cleaned1)
df_cleaned1
```

```python

#Utiliza la siguiente lista de valores que serán considerados como nulos:#
#["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]#

import pandas as pd
import numpy as np

# Lista de valores a considerar como nulos
valores_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]

# Reemplazar  valores por np.nan
df_cleaned1.replace(valores_nulos, np.nan, inplace=True)


print(df_cleaned1)
```

```python
resultado = df_cleaned1.isin(valores_nulos).any().any()

if resultado:
    print("El DataFrame contiene algunos de los valores nulos.")
else:
    print("El DataFrame no contiene ninguno de los valores nulos.")
```

```python
#Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna())#

# Eliminar todas las filas que contienen al menos un NaN
datos_faltantes = df_cleaned1.dropna()

# Mostrar el DataFrame resultante
print(datos_faltantes)
```

```python
hay_nan = datos_faltantes.isna().any().any()

if hay_nan:
    print("El DataFrame contiene valores NaN.")
else:
    print("El DataFrame no contiene valores NaN.")
```

```python
#A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo
#y Máximo. #(hint: Utiliza el método apply sobre la columna.)#


def salarios(Salario_estimado):
    min_salario, max_salario = Salario_estimado.split('-')
    return pd.Series([int(min_salario), int(max_salario)])
datos_faltantes[['Salario_Mínimo', 'Salario_Máximo']] = datos_faltantes['Salary Estimate'].apply(salarios)
print(datos_faltantes)
```

```python

#Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint:#
#utilice reemplazo con diccionario usando el método replace sobre la columna.)#


# Diccionario de recodificación
reemplazo = {
    '10000+': 'Mega Empresas',
    '5001 to 10000 employees': 'Grandes Empresas',
    '1001 to 5000 employees': 'Medianas Empresas',
    '201 to 500 employees': 'Pequeñas Empresas',
    '51 to 200 employees': 'Pequeñas Grandes Empresas',
    '501 to 1000 employees': 'Microempresas',
    'Unknown': 'Empresas sin Información'
}

datos_faltantes['Size'] = datos_faltantes['Size'].replace(reemplazo)

print(datos_faltantes)
```

```python

#Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y#
#la media del salario estimado máximo por tamaño de empresa. (hint: utiliza#
#pd.pivot_table para generar la vista adecuada con las columnas generadas.)#


tabla_pivote = pd.pivot_table(datos_faltantes, 
                              values=['Salario_Mínimo', 'Salario_Máximo'], 
                              index='Size', 
                              aggfunc='mean')

print(tabla_pivote)
```
